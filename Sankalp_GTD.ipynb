```python
import pandas as pd
import numpy as np
import sklearn as sk
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import naive_bayes
from sklearn import ensemble
from sklearn.model_selection import GridSearchCV
from sklearn import neighbors
from sklearn.metrics import accuracy_score

#Loading the data
Terror = pd.read_csv("C:\\Users\\sanka\\Desktop\\Terror3.csv",encoding = 'iso-8859-1',low_memory=False)
Terror.head()
selected_columns = [
    'eventid','iyear','imonth','iday','extended', 'country','region',
    'crit1','crit2','crit3',
    'doubtterr','alternative',
    'success',
    'suicide',
    'attacktype1',
    'targtype1',
    'targsubtype1','guncertain1','individual',
    'claimed','claimmode',
    'weaptype1','weapsubtype1','weaptype3',
    'nkill','nwound',
    'ransom','ID'
]

Terror.drop( 
    [key for key in Terror.keys() if key not in selected_columns] ,
    axis=1, 
    inplace=True)
    

#Obtaining the training and test set based on problem statement
  
success_unknown=Terror.query('ID == 3').copy()
success_known=Terror.query('ID != 3').copy()

success_y = success_known['ID'].as_matrix()
success_X = success_known.drop(['ID'], axis= 1, inplace=False).as_matrix()

#cross Validation

success_X_train, success_X_test, success_y_train, success_y_test = train_test_split(success_X, success_y, test_size=0.2)

#Kneighbours

success_clf = neighbors.KNeighborsClassifier()
success_clf.fit(success_X_train, success_y_train)
success_predictions = success_clf.predict(success_X_test)


print( "Accuracy", accuracy_score(success_y_test, success_predictions) )


#Random Forest

success_clf = ensemble.RandomForestClassifier()
success_clf.fit(success_X_train, success_y_train)
success_predictions = success_clf.predict(success_X_test)

print( "Accuracy", accuracy_score(success_y_test, success_predictions) )        





```
